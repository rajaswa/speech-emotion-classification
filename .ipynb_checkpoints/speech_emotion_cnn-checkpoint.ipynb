Creating a dataframe for training and validation sets

```python
import os
import pandas as pd

categories = ['happy', 'fear', 'disgust', 'sad', 'neutral']
cols = ['file', 'label']

train_df, valid_df = pd.DataFrame(columns=cols), pd.DataFrame(columns=cols)

train_files = []
valid_files = []
train_labels = []
valid_labels = []

for category in categories:
  path = "./meld/train/" + category + "/"
  files = os.listdir(path)
  files = [path + f for f in files]
  train_files = train_files + files
  train_labels = train_labels + [category]*len(files)

for category in categories:
  path = "./meld/val/" + category + "/"
  files = os.listdir(path)
  files = [path + f for f in files]
  valid_files = valid_files + files
  valid_labels = valid_labels + [category]*len(files) 

train_df['file'] = train_files
train_df['label'] = train_labels
valid_df['file'] = valid_files
valid_df['label'] = valid_labels
```

```python
assert(len(train_files)==len(train_labels))
assert(len(valid_files)==len(valid_labels))
```

```python
train_df = train_df.sample(frac=1).reset_index(drop=True)
valid_df = valid_df.sample(frac=1).reset_index(drop=True)
```

# MFCC based Classification

```python
#IMPORTS
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils
import torchvision.models as models
```

```python
label_dict = {'happy':0, 'fear':1, 'disgust':2, 'sad':3, 'neutral':4}
```

Balancing The Dataset

```python
train_df['label'].value_counts()
```

```python
#Balancing Dataset
#Train
happy_df = train_df[train_df['label']=='happy']
fear_df = train_df[train_df['label']=='fear']
disgust_df = train_df[train_df['label']=='disgust']
sad_df = train_df[train_df['label']=='sad']
neutral_df = train_df[train_df['label']=='neutral']

disgust_list = [disgust_df]*3
fear_list = [fear_df]*3
frames = [happy_df[:800], sad_df, neutral_df[:800]] + disgust_list + fear_list 
train_df = pd.concat(frames, ignore_index=True)
train_df.reset_index(inplace=True)

#Valid
happy_df = valid_df[valid_df['label']=='happy']
fear_df = valid_df[valid_df['label']=='fear']
disgust_df = valid_df[valid_df['label']=='disgust']
sad_df = valid_df[valid_df['label']=='sad']
neutral_df = valid_df[valid_df['label']=='neutral']

frames = [happy_df, fear_df, disgust_df, sad_df, neutral_df]
valid_df = pd.concat(frames, ignore_index=True)
valid_df.reset_index(inplace=True)
```

```python
train_df = train_df.sample(frac=1).reset_index(drop=True)
valid_df = valid_df.sample(frac=1).reset_index(drop=True)
```

WRITING CUSTOM DATASET CLASS

```python
class MFCC_Dataset(Dataset):
    """MFCC Spectrogram dataset"""

    def __init__(self, df, transform=None):       
        self.df = df
        self.transform = transform
        self.max_pad_len = 300

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        sample = self.df['file'][idx]
        label = (label_dict[self.df['label'][idx]])
        waveform, sample_rate = torchaudio.load(sample)
        
        '''#MEL Spectrogram
        mel_specgram = torchaudio.transforms.MelSpectrogram(sample_rate=4000, n_mels=40)(waveform)
        mel_specgram = mel_specgram.detach().numpy()
        pad_width = self.max_pad_len - mel_specgram.shape[2] 
        if pad_width > 0:       
          mel_specgram = (np.pad(mel_specgram, pad_width=((0,0), (0,0), (0,pad_width)), mode='constant'))
        else:
          mel_specgram = (mel_specgram[:, :, :self.max_pad_len])'''
        
        #MFCC
        mfcc = torchaudio.transforms.MFCC(sample_rate=8000, log_mels=True, n_mfcc=80)(waveform)
        mfcc = mfcc.detach().numpy()
        pad_width = self.max_pad_len - mfcc.shape[2] 
        if pad_width > 0:       
          mfcc = (np.pad(mfcc, pad_width=((0,0), (0,0), (0,pad_width)), mode='constant'))
        else:
          mfcc = (mfcc[:, :, :self.max_pad_len])

        sample = [mfcc, label]       
        if self.transform:
            sample[0] = self.transform(sample[0])

        return sample
```

Creating Datasets and Dataloaders

```python
train_dataset = MFCC_Dataset(df = train_df)
valid_dataset = MFCC_Dataset(df = valid_df)
```

```python
print(len(train_dataset)) 
print(len(valid_dataset) )
```

```python
train_features = []
train_labels = []

valid_features = []
valid_labels = []
```

```python
for i in range(len(train_dataset)):
  sample = train_dataset[i]
  train_features.append(sample[0])
  train_labels.append(sample[1])
  if i % 500 == 0:
    print(i)

for i in range(len(valid_dataset)):
  sample = valid_dataset[i]
  valid_features.append(sample[0])
  valid_labels.append(sample[1])
  if i % 500 == 0:
    print(i)
```

```python
#RESAMPLING FOR BALANCED DATASET

'''tl = TomekLinks(sampling_strategy='auto')
train_features_res, train_labels_res = tl.fit_resample(train_features, train_labels)

print(len(train_features), len(train_labels))'''
```

```python
train_feature_tensors = torch.stack([torch.Tensor(i) for i in train_features])
train_label_tensors = torch.LongTensor(train_labels)

valid_feature_tensors = torch.stack([torch.Tensor(i) for i in valid_features])
valid_label_tensors = torch.LongTensor(valid_labels)
```

```python
train_tensor_dataset = torch.utils.data.TensorDataset(train_feature_tensors, train_label_tensors)
valid_tensor_dataset = torch.utils.data.TensorDataset(valid_feature_tensors, valid_label_tensors)
```

```python
train_loader = DataLoader(train_tensor_dataset, batch_size=32, shuffle=True, num_workers=1)
valid_loader = DataLoader(valid_tensor_dataset, batch_size=16, shuffle=False, num_workers=1)
```

DEFINING MODEL

```python
# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
# Convolutional neural network 
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(2, 4, kernel_size=3, stride=1, padding=2),
            nn.BatchNorm2d(4),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))
        self.layer2 = nn.Sequential(
            nn.Conv2d(4, 2, kernel_size=3, stride=1, padding=2),
            nn.BatchNorm2d(2),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2))        
        
        self.relu = nn.LeakyReLU()
        self.dropout = nn.Dropout(p=0.2)
        self.fc1 = nn.Linear(3192, 5)
        self.softmax = nn.Softmax()
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)

        out = out.reshape(out.size(0), -1)
        out = self.softmax(self.fc1(out))

        return out
```

```python
# Declaring Model
model = ConvNet().to(device)
```

Setting up weighted loss and optimizer 

```python
# Loss and optimizer
summed = 1609 + 216 + 232 + 705 + 4592
weights = [1609, 216, 232, 705, 4592] 

weights = [x / summed for x in weights]
print(weights)
class_weights = torch.FloatTensor(weights).cuda()
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
# Train the model
train_loss = []
train_accuracy = []
valid_accuracy = []

num_epochs = 15
total_step = len(train_loader)
model.train()
for epoch in range(num_epochs):
    for i, (data, labels) in enumerate(train_loader):
        data = data.to(device)
        labels = labels.to(device)
        
        # Forward pass
        outputs = model(data)
        loss = criterion(outputs, labels)
        
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if (i+1) % 5 == 0:
            print ('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}' 
                   .format(epoch+1, num_epochs, i+1, total_step, loss.item()))
            train_loss.append(loss.item())

            model.eval()  
            correct = 0
            total = 0
            for data, labels in valid_loader:
                data = data.to(device)
                labels = labels.to(device)
                outputs = model(data)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
            print('Valid Accuracy : {} %'.format(100 * correct / total))
            valid_accuracy.append(100 * correct / total)
            model.train()


    model.eval()  
    correct = 0
    total = 0
    for data, labels in train_loader:
      data = data.to(device)
      labels = labels.to(device)
      outputs = model(data)
      _, predicted = torch.max(outputs.data, 1)
      total += labels.size(0)
      correct += (predicted == labels).sum().item()
    print("######################################################")
    print('Train Accuracy : {} %'.format(100 * correct / total))
    print("######################################################")
    train_accuracy.append(100 * correct / total)
    model.train()
```

Final Train & validation Accuracies

```python
model.eval()  
correct = 0
total = 0
for data, labels in valid_loader:
  data = data.to(device)
  labels = labels.to(device)
  outputs = model(data)
  _, predicted = torch.max(outputs.data, 1)
  total += labels.size(0)
  correct += (predicted == labels).sum().item()
print('Valid Accuracy : {} %'.format(100 * correct / total))

correct = 0
total = 0
for data, labels in train_loader:
  data = data.to(device)
  labels = labels.to(device)
  outputs = model(data)
  _, predicted = torch.max(outputs.data, 1)
  total += labels.size(0)
  correct += (predicted == labels).sum().item()
print('Train Accuracy : {} %'.format(100 * correct / total))
```
